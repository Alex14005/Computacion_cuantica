```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros de generacin
N = 50        # Nmero de puntos por rama
K = 3         # Nmero de ramas
mag_noise = 0.2  # Magnitud del ruido
dTheta = 4       # Diferencia angular entre ramas

# Inicializacin de datos
N_train = N * K
x_train = np.zeros((N_train, 2))               # Puntos bidimensionales
y_train = np.zeros(N_train, dtype='uint8')     # Etiquetas (no one-hot)

# Generacin de datos en espiral
for j in range(K):
    ix = range(N * j, N * (j + 1))
    r = np.linspace(0.01, 1, N)  # Radio creciente
    th = np.linspace(j * (2 * np.pi) / K, j * (2 * np.pi) / K + dTheta, N)
    th += np.random.randn(N) * mag_noise       # Agregar ruido a theta
    x_train[ix] = np.c_[r * np.cos(th), r * np.sin(th)]
    y_train[ix] = j

# Visualizacin
plt.figure(figsize=(5, 5))
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, s=40, cmap='viridis')
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.title('Datos en espiral para clasificacin')
plt.grid(True)
plt.show()
```

```python
%matplotlib inline
from IPython import display

import time
import torch

### Feedforward NN sin capa oculta:
class FeedforwardNN(torch.nn.Module):
    def __init__(self, input_size, output_size):
        super(FeedforwardNN, self).__init__()

        # tamaos de capa:
        self.input_size = input_size
        self.output_size = output_size

        # funciones utilizadas dentro del Feedforward NN:
        self.linear1 = torch.nn.Linear(self.input_size, self.output_size)
        self.relu    = torch.nn.ReLU() 
        self.sigmoid = torch.nn.Sigmoid()
        self.softmax = torch.nn.Softmax()      
    def forward(self, x):
        # capa 1:
        linear1_out = self.linear1(x)
        a1 = self.sigmoid(linear1_out)

        # Salida de la red:
        aL = a1

        return aL

### Feedforward NN con una capa oculta:
class FeedforwardNN_withHidden(torch.nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(FeedforwardNN_withHidden, self).__init__()

        # tamaos de capa:
        self.input_size = input_size
        self.hidden_size  = hidden_size
        self.output_size = output_size

        # funciones utilizadas dentro del Feedforward NN:
        self.linear1 = torch.nn.Linear(self.input_size, self.hidden_size)
        self.linear2 = torch.nn.Linear(self.hidden_size, self.output_size)
        self.relu    = torch.nn.ReLU() 
        self.sigmoid = torch.nn.Sigmoid()
        self.softmax = torch.nn.Softmax()      
    def forward(self, x):
        # capa 1:
        linear1_out = self.linear1(x)
        a1 = self.sigmoid(linear1_out)

        # capa 2:
        linear2_out = self.linear2(a1) 
        a2 = self.softmax(linear2_out) 

        # Salida de la red:
        aL = a2

        return aL

input_size = 2
hidden_size = 8
output_size = K
model = FeedforwardNN_withHidden(input_size, hidden_size, output_size)

### Almacenar los datos de entrada como un tensor de PyTorch
x_train = torch.tensor(x_train, dtype = torch.float)

### Codificacin one-hot
y_onehot = np.zeros((y_train.size, K))
y_onehot[np.arange(y_train.size), y_train] = 1
y_onehot = torch.tensor(y_onehot, dtype = torch.float)

### Utilizar la propagacin hacia atrs para minimizar la funcin de costo utilizando el algoritmo de descenso de gradiente:
learning_rate = 3
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)     # SGD = stochastic gradient descent

### Funcin de costo:
cost_func = torch.nn.MSELoss()

N_epochs = 10000 # nmero de veces para ejecutar el descenso de gradiente
```

```python
epoch_list    = []
cost_training = []
acc_training  = []

## Funcin para graficar:
def updatePlot():

    ### Generar coordenadas que cubran todo el plano:
    padding = 0.1
    spacing = 0.02
    
    x1_min, x1_max = x_train[:, 0].min().item() - padding, x_train[:, 0].max().item() + padding
    x2_min, x2_max = x_train[:, 1].min().item() - padding, x_train[:, 1].max().item() + padding
    x1_grid, x2_grid = np.meshgrid(np.arange(x1_min, x1_max, spacing), np.arange(x2_min, x2_max, spacing))

    torch_input = torch.tensor(np.c_[x1_grid.ravel(), x2_grid.ravel()], dtype = torch.float)
    NN_output = model(torch_input)
    predicted_class = np.argmax(NN_output.detach().numpy(), axis=1)

    ### Graficar el clasificador:
    plt.subplot(121)
    plt.contourf(x1_grid, x2_grid, predicted_class.reshape(x1_grid.shape), K, alpha=0.8)
    plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, s=40)
    plt.xlim(x1_grid.min(), x1_grid.max())
    plt.ylim(x2_grid.min(), x2_grid.max())
    plt.xlabel(r'$x_1$')
    plt.ylabel(r'$x_2$')

    ### Graficar la funcin de costo durante el entrenamiento:
    plt.subplot(222)
    plt.plot(epoch_list, cost_training, 'o-')
    plt.xlabel('')
    plt.ylabel('Costo de entrenamiento')

    ### Graficar la precisin del entrenamiento:
    plt.subplot(224)
    plt.plot(epoch_list, acc_training, 'o-')
    plt.xlabel('Epoca')
    plt.ylabel('Precisin de entrenamiento')
    

### Entrenar durante varias pocas: ###
for epoch in range(N_epochs):

    optimizer.zero_grad()                   # establecer los gradientes en cero (necesario ya que PyTorch acumule los gradientes)
    NN_output = model(x_train)              # Salida de la red neuronal
    cost = cost_func(NN_output, y_onehot)
    cost.backward()                         # calcula los gradientes
    optimizer.step()                        # actualizando los parametros

    ### Actualizar la grfica e imprimir los resultados cada 500 pocas:
    if epoch % 500 == 0:
        predicted_class = np.argmax(NN_output.detach().numpy(), axis=1)
        accuracy = np.mean(predicted_class == y_train)
        
        epoch_list.append(epoch)
        cost_training.append(cost.detach().numpy())
        acc_training.append(accuracy)
        
        ### Actualizar la grfica del clasificador resultante:
        plt.close(2)
        fig = plt.figure(2, figsize=(10,5))
        #fig.subplots_adjust(hspace=.3, wspace=.3)
        plt.clf()
        updatePlot()
        display.display(fig)
        print("Iteracin %d:\n  Costo de entrenamiento %f\n  Precisin de entrenamiento %f\n" % (epoch, cost, accuracy) )
        display.clear_output(wait=True)
        time.sleep(0.2) # para reducir la velocidad de las actualizaciones de la grfica

print("Costo final de entrenamiento %f\nPrecisin final de entrenamiento %f\n" % (cost, accuracy) )
```

```python
## Ejemplo sencillo variacional

from qiskit import QuantumCircuit
import qiskit.quantum_info as qi
from qiskit.visualization import plot_bloch_multivector

import numpy as np
from math import radians, degrees
from scipy.optimize import minimize

def costo_edo_1(proba_edo_1):
    return 1 - proba_edo_1 # diferencia de probabilidad de que sea 1

def fn_objetivo(angle, fn_costo, shots=2000):
    qc = QuantumCircuit(1)

    qc.ry(radians(angle[0]), 0) # Rotacion en Y por una cantidad de 'angle' grados

    sv = qi.Statevector(qc)
    counts = sv.sample_counts(shots=shots)

    proba_edo_1 = (counts['1'] if '1' in counts else 0)/shots # probabilidad de que el estado sea 1

    cost = fn_costo(proba_edo_1)

    print('Intento: angulo =', angle, 'produce un costo promedio =', cost)

    return cost
```

```python
# Definimos el valor inicial para el parametro que estamos buscando
initial_guess: int = 90
```

```python
# Usamos la funcion minimize del paquete scipy para encontrar el valor del angulo que estamos buscando
result = minimize(fn_objetivo, initial_guess, args=(costo_edo_1,), method='Powell')
```

```python
print('/n Angulo optimizado en grados: ', [a %360 for a in result.x])
```

```python
#Mostramos el estado que se logra con el parametro encontrado en una
#Esfera de Bloch

qc = QuantumCircuit(1)
qc.ry(radians(result.x[0]), 0)

sv = qi.Statevector(qc)
plot_bloch_multivector(sv)
```

```python
# Ahora encontramos el valor del angulo necesario para terminar en el estado |-i>

def costo_edo_menos_y(average_state: float) -> float:
    return (average_state - 0.5)**2

initial_guess = 0
result = minimize(fn_objetivo, initial_guess, args=(costo_edo_menos_y,), method='Powell')
```

```python
qc = QuantumCircuit(1)
qc.rx(radians(result.x[0]), 0)

sv = qi.Statevector(qc)
plot_bloch_multivector(sv)
```
