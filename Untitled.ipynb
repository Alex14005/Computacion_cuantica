```python
from qiskit import QuantumRegister

# Este tendra en numero por defecto
qubit1 = QuantumRegister(1)
#Aqui se le pone nombre al q2
qubit2 = QuantumRegister(1, 'qreg')
```

```python
print(qubit1)
print(qubit2)
```

```python
from qiskit import QuantumCircuit

circuit = QuantumCircuit(qubit1)

circuit.draw('mpl') #'mpl' es matplotlib
```

```python
circ = QuantumCircuit(3, 2) #(cuantos qbits y cuantos bits)

circ.h(0) # aplicar compuerta H (hadamart)
circ.cx(0, 1)
circ.cx(0, 2)
circ.measure(0, 1)
circ.measure(2, 0)

circ.draw('mpl')
```
